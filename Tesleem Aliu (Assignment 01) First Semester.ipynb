## Question 1: Mention 5 simple data types used in Python and state specific use cases for each

### 5 simple data types and their use cases:
- Integer(int); This is used to represent whole numbers in python programming, such as 4, 7, -20, 0, etc.



- String(str); This is used to represent or store texts and characters including, letters, digits, and symbols enclosed in a single or double quotation mark in python programming. Such as; 'Altschool Africa', "School of Data 2023 Tinyuka".



- Boolean(bool); This is used to represent 2 values known as True or False. These values are often used in conditional statements in python programming.



- Float(float); This is used to decimal numbers in python programming. Such as; 3.20, 7.7, 0.5, etc.



- List; This is used in python programming to store and manage collections of different items, which can also include duplicate values. Such as [1, 'Gideon', 2, 'Susan', 3, 'Sandrine']

## Question 2: Mention 4 data structures used in Python and state specific use cases for each.

### 4 data structrues and their use cases:
- Dictionaries; These are key-value pairs used to map and associate values with unique keys. Such as {'Instructor': 'Gideon', 'Program Manager': 'Susan', 'Community Associate': 'Sandrine'},  {'School of Data': ['Data Science', 'Data Engineering', 'Data Analysis'], 'School of Engineering': ['Backend Engineering', 'Frontend Engineering', 'Cloud Engineering'], 'School of Product': ['Product Management', 'Product Design', 'Product Marketing']}



- Stacks; These are used to apply the last-in-first-out (LIFO) behaviour in python programming. They are commonly used for tasks like undo and redo operations,  backtracking algorithms, memory management, etc.



- Queues; These are used to implement the first-in-first-out (FIFO) behaviour in python programming. They are used in task where items are processed in a specific order, such as print queue, task scheduling, call center systems, etc.



- Hash Tables; These are used to provide fast and efficient way to retrieve and store data in a key-value pair format. They are used extensively in implementing dictionaries and databases.
