1. From the list (monthly_sales) above, 
- create four seperate lists frm the list for each quarter namely: q1, q2, q3, q4. 
- report what quarter had the highest amount in total revenue, 
- report the lowest.

```python
# List of month_sales
monthly_sales = [120, 135, 150, 145, 155, 160, 170, 180, 190, 200, 210, 220]

# List for each quarter q1, q2, q3, q4
q1 = monthly_sales[:3]
q2 = monthly_sales[3:6]
q3 = monthly_sales[6:9]
q4 = monthly_sales[9:12]

# Creating a dictionary for total revenue of each quarter
total_quarterly_revenue = {
    'Total_revenue_in_q1': sum(q1),
    'Total_revenue_in_q2': sum(q2),
    'Total_revenue_in_q3': sum(q3),
    'Total_revenue_in_q4': sum(q4)
}

# Print total revenue for each quarter
print(total_quarterly_revenue)

# The quarter with the highest amount of revenue
print(f'The quarter with the highest amount in total revenue is {max(total_quarterly_revenue)};' 
      f'the total revenue for that quarter is {max(total_quarterly_revenue.values())}')

# The quarter with the lowest amount of revenue
print(f'The quarter with the lowest amount in total revenue is {min(total_quarterly_revenue)};' 
      f'the total revenue for that quarter is {min(total_quarterly_revenue.values())}')
```

3. You are managing a list of students' names in your class. 
* Create a list called student_names containing the following student names as strings: 'Alice', 'Bob', 'Charlie', 'David', 'Eve'.
- Use the student_names list from the list above for the following operations:
  - a. Print the number of students in the class (the length of the list).
  - b. Print the first and last student names in the list.
  - c. Charlie has left the class. Remove his name from the list.
  - d. A new student, "Frank," has joined the class. Add his name to the list.
  - e. The class has been dissolved, and you need to remove all student names from the list.

```python
# Creating a list of Students name
student_names = ['Alice', 'Bob', 'Charlie', 'David', 'Eve']

# Number of students in the class
print(f'The number of students in the class are {len(student_names)}')

# First and last student in the list
print(f'The first student in the list is {student_names[0]}, ' 
      f'while the last student in the list is {student_names[-1]}')

# Removing Charlie from the list 
student_names.remove('Charlie')
print(student_names)

# Adding Frank to the list
student_names.append('Frank')
print(student_names)

# Removing all students from the list
student_names.clear()
print(student_names)
```
